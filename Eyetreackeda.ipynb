```python
import nbformat as nbf
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
#import plotly.express as px
```

```python
df = pd.read_csv('eye_tracking_data.csv', sep=';')
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
print('1. Barplot of Gender Distribution')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')
plt.show()
```

```python
print('2. Barplot of Native Language')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Nativelang')
plt.title('Native Language(Yes/No)')
plt.show()
```

```python
print('3. Barplot of Other Language')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Otherlang')
plt.title('Other Language(Yes/No)')
plt.show()
```

```python
print('4. Histogram of Age Distribution')
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
print('5. Bar plot of Dyslexia count')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Dyslexia')
plt.title('Dyslexia Count (Yes/No)')
plt.show()
```

```python
print('6. Boxplot of Age by Dyslexia Status')
plt.figure(figsize=(10, 6)) 
sns.boxplot(data=df, x='Dyslexia', y='Age')
plt.title('Age Distribution by Dyslexia Status')
plt.show()
```

```python
print('7. Stacked bar of gender by dyslexia')
plt.figure(figsize=(10, 6))
pd.crosstab(df['Gender'], df['Dyslexia']).plot(kind='bar', stacked=False)
plt.title('Gender Distribution by Dyslexia Status')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('8. Stacked bar of native language by dyslexia')
plt.figure(figsize=(10, 6))
pd.crosstab(df['Nativelang'], df['Dyslexia']).plot(kind='bar', stacked=False)
plt.title('Native Language Distribution by Dyslexia Status')
plt.xlabel('Native Language (Yes/No)')
plt.ylabel('Count')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('9. Stacked bar of other language by dyslexia')
plt.figure(figsize=(10, 6))
pd.crosstab(df['Otherlang'], df['Dyslexia']).plot(kind='bar', stacked=False)
plt.title('Other Language Distribution by Dyslexia Status')
plt.xlabel('Other Language (Yes/No)')
plt.ylabel('Count')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('10. Line plot of mean accuracy across blocks (1-32) for dyslexia groups')
acc_cols = [f'Accuracy{i}' for i in range(1, 33)]
df_acc = df.melt(id_vars=['Dyslexia'], value_vars=acc_cols, var_name='Block', value_name='Accuracy')
df_acc['Block'] = df_acc['Block'].str.extract('(\d+)').astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_acc, x='Block', y='Accuracy', hue='Dyslexia', errorbar='se')
plt.title('Mean Accuracy Across Blocks by Dyslexia Status')
plt.xlabel('Block')
plt.ylabel('Mean Accuracy')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('11. Line plot of mean missarate across blocks for dyslexia groups')
miss_cols = [f'Missrate{i}' for i in range(1, 33)]
df_miss = df.melt(id_vars=['Dyslexia'], value_vars=miss_cols, var_name='Block', value_name='Missrate')
df_miss['Block'] = df_miss['Block'].str.extract('(\d+)').astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_miss, x='Block', y='Missrate', hue='Dyslexia', errorbar='se')
plt.title('Mean Missrate Across Blocks by Dyslexia Status')
plt.xlabel('Block')
plt.ylabel('Mean Missrate')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('12. Line plot of mean hits across blocks for dyslexia groups')
hits_cols = [f'Hits{i}' for i in range(1, 33)]
df_hits = df.melt(id_vars=['Dyslexia'], value_vars=hits_cols, var_name='Block', value_name='Hits')
df_hits['Block'] = df_hits['Block'].str.extract('(\d+)').astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_hits, x='Block', y='Hits', hue='Dyslexia', errorbar='se')
plt.title('Mean Hits Across Blocks by Dyslexia Status')
plt.xlabel('Block')
plt.ylabel('Mean Hits')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('13. Line plot of mean Clicks across blocks for dyslexia groups')
clicks_cols = [f'Clicks{i}' for i in range(1, 33)]
df_clicks = df.melt(id_vars=['Dyslexia'], value_vars=clicks_cols, var_name='Block', value_name='Clicks')
df_clicks['Block'] = df_clicks['Block'].str.extract('(\d+)').astype(int)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_clicks, x='Block', y='Clicks', hue='Dyslexia', errorbar='se')
plt.title('Mean Clicks Across Blocks by Dyslexia Status')
plt.xlabel('Block')
plt.ylabel('Mean Clicks')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('14.Heatmap of correlation between Accuracy across blocks')
acc10 = [f'Accuracy{i}' for i in range(1, 11)]
corr = df[acc10].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', center=0)
plt.title('Correlation Heatmap of Accuracy Across Blocks 1-10')
plt.show()
```

```python
print('15. Boxplot of overall mean Accuracy per person (averaged across blocks) by Dyslexia')
df['MeanAccuracy'] = df[acc_cols].mean(axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Dyslexia', y='MeanAccuracy')
plt.title('Mean Accuracy by Dyslexia Status')
plt.xlabel('Dyslexia (Yes/No)')
plt.ylabel('Mean Accuracy')
plt.show()
```

```python
print('16. Boxplot of overall mean Missrate per person by Dyslexia')
df['MeanMissrate'] = df[miss_cols].mean(axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Dyslexia', y='MeanMissrate')
plt.title('Mean Missrate by Dyslexia Status')
plt.xlabel('Dyslexia (Yes/No)')
plt.ylabel('Mean Missrate')
plt.show()  
```

```python
print('17. Boxplot of overall mean Hits per person by Dyslexia')
df['MeanHits'] = df[hits_cols].mean(axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Dyslexia', y='MeanHits')
plt.title('Mean Hits by Dyslexia Status')
plt.xlabel('Dyslexia (Yes/No)')
plt.ylabel('Mean Hits')
plt.show()
```

```python
print('18. Boxplot of overall mean Clicks per person by Dyslexia')
df['MeanClicks'] = df[clicks_cols].mean(axis=1)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Dyslexia', y='MeanClicks')
plt.title('Mean Clicks by Dyslexia Status')
plt.xlabel('Dyslexia (Yes/No)')
plt.ylabel('Mean Clicks')
plt.show()
```

```python
print('19. Scatter plot of Age vs overall mean Accuracy, colored by Dyslexia')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='MeanAccuracy', hue='Dyslexia', alpha=0.7)
sns.regplot(data=df[df['Dyslexia'] == 'Yes'], x='Age', y='MeanAccuracy', scatter=False, label='Dyslexia: Yes', color='blue')
sns.regplot(data=df[df['Dyslexia'] == 'No'], x='Age', y='MeanAccuracy', scatter=False, label='Dyslexia: No', color='orange')
plt.title('Age vs Mean Accuracy by Dyslexia Status')
plt.xlabel('Age')
plt.ylabel('Mean Accuracy')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('20. VIolin plot of Age by Dyslexia')
plt.figure(figsize=(12, 10))
sns.violinplot(data=df, x='Dyslexia', y='Age', inner='quartile')
plt.title('Age Distribution by Dyslexia Status')
plt.xlabel('Dyslexia (Yes/No)')
plt.ylabel('Age')
plt.show()
```

```python
print('21. Violin plot of Age by Dyslexia')
sns.pairplot(data=df[['Age', 'MeanAccuracy', 'MeanMissrate', 'MeanHits', 'MeanClicks', 'Dyslexia']], hue='Dyslexia', diag_kind='kde')
plt.suptitle('Pairplot of Age, Mean Accuracy, Missrate, Hits, Clicks by Dyslexia Status', y=1.02)
plt.show()  
```

```python
print('22. Count plot of Dyslexia by Gender and Native  Language')
ct = pd.crosstab([df['Gender'], df['Nativelang']], df['Dyslexia'])
ct.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.title('Dyslexia Count by Gender and Native Language')
plt.xlabel('Gender and Native Language')
plt.ylabel('Count')
plt.legend(title='Dyslexia')
plt.show()
```

```python
print('23. A histogram Distribution of total Clicks (sum across blocks) by Dyslexia ')
df['TotalClicks'] = df[clicks_cols].sum(axis=1)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TotalClicks', hue='Dyslexia', bins=20, kde=True)
plt.title('Distribution of Total Clicks by Dyslexia Status')
plt.xlabel('Total Clicks (sum across blocks)')
plt.ylabel('Count') 
plt.show()
```

```python
print('24. Distribution of total Hits (sum across blocks) by Dyslexia')
df['TotalHits'] = df[hits_cols].sum(axis=1)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TotalHits', hue='Dyslexia', bins=20, kde=True)
plt.title('Distribution of Total Hits by Dyslexia Status')
plt.xlabel('Total Hits (sum across blocks)')    
plt.ylabel('Count')
plt.show()
```

```python
print('25. Distribution of total Misses (sum across blocks) by Dyslexia')
miss_cols_num = [f'Misses{i}' for i in range(1, 33)]
df['TotalMisses'] = df[miss_cols_num].sum(axis=1)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TotalMisses', hue='Dyslexia', bins=20, kde=True)
plt.title('Distribution of Total Misses by Dyslexia Status')
plt.xlabel('Total Misses (sum across blocks)')  
plt.ylabel('Count') 
plt.show()
```
